```python
# -*- coding: utf-8 -*-
"""
CCA demo with original data in 2D or 3D, reduced to 1D or 2D respectively.
Includes explicit U1 vs V1 and U2 vs V2 plots with correlation annotations.
- Uses sample-covariance convention Cov(U,V) = (1/(n-1)) U^T V
- Whitened CCA (with optional ridge)
Author: ChatGPT (updated)
"""

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting

# -------------------------
# Utility: symmetric power
# -------------------------
def symmetric_matrix_power(M, power, eps=1e-12):
    vals, vecs = np.linalg.eigh(M)
    idx = np.argsort(vals)[::-1]
    vals = vals[idx]
    vecs = vecs[:, idx]
    if power < 0:
        vals = np.clip(vals, a_min=eps, a_max=None)
    Dp = np.diag(np.power(vals, power))
    return vecs @ Dp @ vecs.T

# -------------------------
# CCA (whitened SVD) function
# -------------------------
def cca(X, Y, n_components=None, reg=0.0, eps=1e-12):
    n = X.shape[0]
    assert X.shape[0] == Y.shape[0], "X and Y must have same n"
    Sigma_XX = (X.T @ X) / (n - 1)
    Sigma_YY = (Y.T @ Y) / (n - 1)
    Sigma_XY = (X.T @ Y) / (n - 1)
    Sigma_XX_reg = Sigma_XX + reg * np.eye(Sigma_XX.shape[0])
    Sigma_YY_reg = Sigma_YY + reg * np.eye(Sigma_YY.shape[0])
    Sxx_inv_sqrt = symmetric_matrix_power(Sigma_XX_reg, -0.5, eps=eps)
    Syy_inv_sqrt = symmetric_matrix_power(Sigma_YY_reg, -0.5, eps=eps)
    R = Sxx_inv_sqrt @ Sigma_XY @ Syy_inv_sqrt
    U, S, VT = np.linalg.svd(R, full_matrices=False)
    V = VT.T
    max_r = min(R.shape[0], R.shape[1])
    r = max_r if n_components is None else min(n_components, max_r)
    rhos = S[:r]
    A = Sxx_inv_sqrt @ U[:, :r]
    B = Syy_inv_sqrt @ V[:, :r]
    U_proj = X @ A
    V_proj = Y @ B
    return A, B, rhos, U_proj, V_proj

def sample_cov(U, V):
    n = U.shape[0]
    return (U.T @ V) / (n - 1)

# -------------------------
# Data generator: shared latent factors
# -------------------------
def generate_xy_shared_latent(n, dim_x, dim_y, k_latent=1, snr=5.0, random_state=None):
    rng = np.random.RandomState(random_state)
    Z = rng.normal(size=(n, k_latent))  # latent factors
    Wx = rng.normal(scale=1.0, size=(dim_x, k_latent))
    Wy = rng.normal(scale=1.0, size=(dim_y, k_latent))
    X_signal = Z @ Wx.T
    Y_signal = Z @ Wy.T
    var_X_signal = np.mean(X_signal**2)
    var_Y_signal = np.mean(Y_signal**2)
    noise_var_X = var_X_signal / snr
    noise_var_Y = var_Y_signal / snr
    X = X_signal + rng.normal(scale=np.sqrt(noise_var_X), size=(n, dim_x))
    Y = Y_signal + rng.normal(scale=np.sqrt(noise_var_Y), size=(n, dim_y))
    X -= X.mean(axis=0)
    Y -= Y.mean(axis=0)
    return X, Y, Z, Wx, Wy

# -------------------------
# Plot raw dataset (2D or 3D)
# -------------------------
def plot_raw(X, Y, title="Raw data (X left, Y right)"):
    dim_x = X.shape[1]
    dim_y = Y.shape[1]
    plt.figure(figsize=(10,4))
    ax1 = plt.subplot(1,2,1, projection='3d') if dim_x==3 else plt.subplot(1,2,1)
    if dim_x == 2:
        ax1.scatter(X[:,0], X[:,1], alpha=0.7)
        ax1.set_xlabel("X dim 1"); ax1.set_ylabel("X dim 2")
    elif dim_x == 3:
        ax1.scatter(X[:,0], X[:,1], X[:,2], alpha=0.7)
        ax1.set_xlabel("X dim 1"); ax1.set_ylabel("X dim 2"); ax1.set_zlabel("X dim 3")
    else:
        ax1.plot(X[:,0], np.zeros_like(X[:,0]), 'o')
        ax1.set_xlabel("X dim 1")
    ax1.set_title("Raw X ({}D)".format(dim_x))
    ax2 = plt.subplot(1,2,2, projection='3d') if dim_y==3 else plt.subplot(1,2,2)
    if dim_y == 2:
        ax2.scatter(Y[:,0], Y[:,1], alpha=0.7)
        ax2.set_xlabel("Y dim 1"); ax2.set_ylabel("Y dim 2")
    elif dim_y == 3:
        ax2.scatter(Y[:,0], Y[:,1], Y[:,2], alpha=0.7)
        ax2.set_xlabel("Y dim 1"); ax2.set_ylabel("Y dim 2"); ax2.set_zlabel("Y dim 3")
    else:
        ax2.plot(Y[:,0], np.zeros_like(Y[:,0]), 'o')
        ax2.set_xlabel("Y dim 1")
    ax2.set_title("Raw Y ({}D)".format(dim_y))
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# -------------------------
# Plot reduced embeddings
# -------------------------
def plot_reduced(U_proj, V_proj, rhos, reduce_dim, title_suffix=""):
    if reduce_dim == 1:
        plt.figure(figsize=(12,4))
        plt.subplot(1,3,1)
        plt.scatter(U_proj[:,0], V_proj[:,0], alpha=0.7)
        plt.xlabel("U1 (X a1)"); plt.ylabel("V1 (Y b1)")
        plt.title(f"U1 vs V1 (rho{rhos[0]:.3f})")
        plt.grid(True)
        plt.subplot(1,3,2)
        plt.hist(U_proj[:,0], bins=30, alpha=0.7)
        plt.title("Histogram U1")
        plt.subplot(1,3,3)
        plt.hist(V_proj[:,0], bins=30, alpha=0.7)
        plt.title("Histogram V1")
        plt.suptitle("Reduced (1D) " + title_suffix)
        plt.tight_layout()
        plt.show()
    elif reduce_dim == 2:
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1)
        plt.scatter(U_proj[:,0], U_proj[:,1], alpha=0.7)
        plt.xlabel("U1"); plt.ylabel("U2"); plt.title("X -> U (2D)")
        plt.grid(True)
        plt.subplot(1,2,2)
        plt.scatter(V_proj[:,0], V_proj[:,1], alpha=0.7)
        plt.xlabel("V1"); plt.ylabel("V2"); plt.title("Y -> V (2D)")
        plt.grid(True)
        plt.suptitle("Reduced (2D) " + title_suffix)
        plt.tight_layout()
        plt.show()
    else:
        raise ValueError("reduce_dim must be 1 or 2")

# -------------------------
# Plot U1 vs V1 and U2 vs V2 with corr annotations
# -------------------------
def plot_component_correlations(U_proj, V_proj, rhos, comps=(1,2), title_suffix=""):
    """
    Plot scatter plots for component pairs (i,i) for i in comps if available.
    Shows Pearson correlation and canonical rho in the title.
    comps is 1-based indexing for readability.
    """
    available = min(U_proj.shape[1], V_proj.shape[1])
    to_plot = [i for i in comps if i <= available]
    if not to_plot:
        return
    n = len(to_plot)
    plt.figure(figsize=(5*n,4))
    for idx, comp in enumerate(to_plot, start=1):
        i = comp - 1  # zero-based
        x = U_proj[:, i]
        y = V_proj[:, i]
        # Pearson correlation
        corr = np.corrcoef(x, y)[0,1]
        plt.subplot(1, n, idx)
        plt.scatter(x, y, alpha=0.7)
        plt.xlabel(f"U{comp}"); plt.ylabel(f"V{comp}")
        # Annotate with numeric values
        rho = rhos[i] if i < len(rhos) else None
        plt.title(f"U{comp} vs V{comp}: Pearson r={corr:.3f}, rho={rho:.3f}" if rho is not None else f"U{comp} vs V{comp}: r={corr:.3f}")
        plt.grid(True)
    plt.suptitle("Component-wise correlation " + title_suffix)
    plt.tight_layout()
    plt.show()

# -------------------------
# Main / usage
# -------------------------
if __name__ == "__main__":
    # Settings
    original_dim = 3   # choose 2 or 3
    dim_x = original_dim
    dim_y = original_dim
    reduce_dim = 1 if original_dim==2 else 2
    n = 300
    k_latent = 1 if reduce_dim==1 else 2
    snr = 8.0
    reg = 1e-6

    # generate data
    X, Y, Z, Wx, Wy = generate_xy_shared_latent(n=n, dim_x=dim_x, dim_y=dim_y,
                                                k_latent=k_latent, snr=snr, random_state=42)

    # plot raw
    plot_raw(X, Y, title=f"Raw original data (dim={original_dim})")

    # run CCA
    A, B, rhos, U_proj_full, V_proj_full = cca(X, Y, n_components=reduce_dim, reg=reg)
    print(f"Top {reduce_dim} canonical correlations: {rhos[:reduce_dim]}")

    # verify sample covariance
    C = sample_cov(U_proj_full, V_proj_full)
    print("Sample covariance matrix between canonical variates (1/(n-1) U^T V):\n", C)

    # plot reduced embeddings
    plot_reduced(U_proj_full, V_proj_full, rhos, reduce_dim,
                 title_suffix=f"(orig dim={original_dim} -> reduced dim={reduce_dim})")

    # plot component-wise correlations (U1 vs V1, U2 vs V2)
    plot_component_correlations(U_proj_full, V_proj_full, rhos, comps=(1,2),
                                title_suffix=f"(orig dim={original_dim} -> reduced dim={reduce_dim})")

    # optional: time-series alignment plot for first component (helps visualize)
    plt.figure(figsize=(8,3))
    plt.plot(U_proj_full[:,0], label='U1 (X a1)')
    plt.plot(V_proj_full[:,0], label='V1 (Y b1)')
    plt.legend()
    plt.title('First canonical variates (time-series view)')
    plt.tight_layout()
    plt.show()
```
